# Dataset description

The dataset reports the percentage of the population with access to electricity across different countries worldwide. This value is one of the key international indicators used to assess a country’s level of energy development.

The information comes from institutional sources such as the World Bank and the Sustainable Energy for All (SE4ALL) initiative, which compile data from industries, energy agencies, national surveys, and other official bodies. The dataset is also linked to the themes of Sustainable Development Goal 7 (SDG 7), which aims to ensure modern, sustainable, reliable, and affordable energy for all.

```python
energy['Mean_2012_2015'] = energy[['2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]']].apply(lambda row: pd.to_numeric(row, errors='coerce').mean(), axis=1)
```

# 5. Graphs

Import the matplotlib and seaborn libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

* **Matplotlib**

# Histogram
Shows the distribution of energy consumption in 2016 across all countries, highlighting which ranges are most common

```python
plt.figure(figsize=(8,5))
plt.hist(energy['2016 [YR2016]'], bins=15, color='orange', edgecolor='black')
plt.title('Distribution of Energy Consumption in 2016')
plt.xlabel('Energy Consumption (2016)')
plt.ylabel('Number of Countries')

plt.xticks(rotation=45)
plt.show()
```

This histogram shows how the 2016 energy consumption values are distributed across all countries. Most countries fall into the lower ranges, while only a few reach very high consumption levels. The distribution is heavily skewed, meaning that high-consumption countries are rare compared to low- and mid-level ones.

# Scatterplot
Focuses on Asian countries to compare energy consumption at the country level in 2016

```python
plt.figure(figsize=(8,5))
plt.scatter(energy[energy['Continent']=='Asia']['Country Name'], energy[energy['Continent']=='Asia']['2016 [YR2016]'], color='orange', s=100)
plt.title('Energy Consumption in 2016 - Asian Countries')
plt.xlabel('Country')
plt.ylabel('Energy Consumption (2016)')
plt.xticks(rotation=90)
plt.show()
```

This scatterplot highlights the differences within Asia in 2016. Each country shows a specific consumption level, and the plot makes it clear that Asian countries vary widely. Some nations consume very little energy, while others reach values close to the global maximum. This helps identify outliers and compare countries directly.

* **Seaborn**

# Boxplot
Compares energy consumption between Asia and Europe in 2016, providing a summary of central tendencies and variability

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Continent', y='2016 [YR2016]', data=energy[ (energy['Continent'] == 'Asia') | (energy['Continent'] == 'Europe')])
plt.title('Energy Consumption in 2016: Asia vs Europe')
plt.xlabel('Continent')
plt.ylabel('Energy Consumption (2016)')
plt.show()
```

This boxplot compares the overall distribution between Asia and Europe. Europe shows a narrower range, indicating more stable and similar values across countries. Asia, instead, displays greater variability and includes both very low and very high consumption levels. This demonstrates a higher internal diversity within the Asian continent.

# Heatmap
Displays the correlation of energy consumption across the years 2012–2016, helping identify trends over time

```python
plt.figure(figsize=(8,6))
sns.heatmap(energy[['2012 [YR2012]','2013 [YR2013]','2014 [YR2014]','2015 [YR2015]','2016 [YR2016]']].replace('..',0).astype(float).corr(), annot=True)
plt.title('Correlation of Energy Consumption (2012-2016)')
plt.show()
```

The heatmap shows that the yearly energy consumption values are strongly correlated with each other. High correlations (close to 1) indicate that countries tend to maintain similar consumption patterns over time. This suggests consistency and gradual evolution rather than sudden changes across the years.

# Stripplot
Shows individual country data points for Asia vs Europe in 2016, highlighting the spread and outliers for each continent

```python
plt.figure(figsize=(8,6))
sns.stripplot(x='Continent', y='2016 [YR2016]', data=energy[energy['Continent'].isin(['Asia','Europe'])].replace('..',0).astype({'2016 [YR2016]': float}), jitter=True, hue='Continent')
plt.title('Energy Consumption in 2016: Asia vs Europe (individual data points)')
plt.xlabel('Continent')
plt.ylabel('Energy Consumption 2016')
plt.show()
```

This stripplot displays every single country’s 2016 consumption in Asia and Europe. It confirms the findings from the boxplot: Europe has a compact and stable cluster of values, while Asia includes countries both very low and very high on the scale. The plot helps visualize the spread and identify individual positions within each continent.

*Note*

1. **`pd.to_numeric(..., errors='coerce')`** = to convert the column values to numeric, turning any non-numeric strings (like `'..'`) into `NaN`, so that the mean can be calculated without errors

2. **`.replace('..', 0)`** = replace `'..'` with `0` because `'..'` represents missing or unavailable data in the dataset. Replacing it with a numeric value allows to perform calculations without errors

3. **`.astype(float)`** = convert the column values to `float` so that numeric operations (like correlation) can be performed, since the original data is stored as strings

4. **`.corr()`** = use `.corr()` to calculate the correlation between numeric columns. This shows how energy consumption in different years or regions is related to each other
